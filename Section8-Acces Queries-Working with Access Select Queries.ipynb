```sql
SELECT OrderID, OrderDate, RequiredDate, ShipDate
FROM Orders
WHERE [ShipDate] >= [RequiredDate];




# Resumen: Usar comodines (wildcards) en criterios de consultas de Access con `LIKE`

**Idea clave:** En Microsoft Access puedes filtrar texto usando **comodines** con el operador `LIKE`. Esto te permite encontrar valores que **empiezan**, **terminan** o **contienen** ciertos caracteres o secuencias, sin tener que escribir el valor exacto.

## Comodines principales (modo ANSI-89 predeterminado de Access)
- `*` ‚Üí ‚Äúcualquier cadena de cualquier longitud‚Äù (incluye vac√≠o).
- `?` ‚Üí ‚Äúun √∫nico car√°cter‚Äù.

> Nota: Si tu base est√° en **ANSI-92**, los comodines cambian a `%` (cualquier cadena) y `_` (un car√°cter). En el dise√±ador de Access, por defecto se usan `*` y `?`.

## Ejemplos de criterios en la **fila Criterios** (Vista de dise√±o)

1. **Empieza con una letra**  
   - En el campo *Ship Country* (Pa√≠s de env√≠o):  
     ```
     Like "S*"
     ```
   - Devuelve pa√≠ses que **empiezan** con `S` (Spain, Sweden, Switzerland, etc.).

2. **Termina con una letra**  
Like "*n"

- Devuelve valores que **terminan** con `n` (e.g., Spain, Sweden).

3. **Contiene una letra o secuencia en cualquier posici√≥n**  
Like "s"

- Devuelve valores que **contienen** la letra `s` en cualquier parte (Austria, Switzerland, USA, Sweden, etc.).
- Para una **secuencia** concreta:  
  ```
  Like "*land"
  ```
  - Termina en `land` (Switzerland, Finland, Ireland).

4. **Secuencia no necesariamente al inicio/fin**  
- Contiene la secuencia `sw` en cualquier parte (consecutiva):  
  ```
  Like "*sw*"
  ```
- Contiene una `S` y **luego** una `W` (no necesariamente juntas):  
  ```
  Like "S*W*"
  ```
  > Si quieres ignorar may√∫sculas/min√∫sculas, Access normalmente ya lo hace (ver notas).

5. **Rangos por inicial (√∫til para apellidos o clasificaciones alfab√©ticas)**
- Empieza con A‚ÄìG (aprox. usando conjuntos):  
  ```
  Like "[A-G]*"
  ```
- No empieza con vocal:
  ```
  Not Like "[AEIOU]*"
  ```

## Pasos r√°pidos en la Vista de dise√±o
1. Abre la consulta en **Vista de dise√±o**.
2. En la columna del campo de texto (por ejemplo, **Ship Country**), escribe el criterio con comodines (por ejemplo, `Like "S*"`).
3. Al salir del campo, Access puede **insertar autom√°ticamente** `Like` si escribiste solo el patr√≥n (por ejemplo, `S*`).
4. Ejecuta la consulta para ver los resultados.

## Ejemplo de SQL
```sql
-- Empieza con S
SELECT OrderID, ShipCountry
FROM Orders
WHERE ShipCountry Like "S*";

-- Termina con n
SELECT OrderID, ShipCountry
FROM Orders
WHERE ShipCountry Like "*n";

-- Contiene 'land'
SELECT OrderID, ShipCountry
FROM Orders
WHERE ShipCountry Like "*land";


### Notas y buenas pr√°cticas

May√∫sculas/min√∫sculas: Por defecto, Like en Access suele ser no sensible a may√∫sculas (depende de la configuraci√≥n regional/Collation). ‚ÄúS‚Äù y ‚Äús‚Äù suelen equivaler.
ANSI-89 vs ANSI-92: Si tu base usa ANSI-92, usa % y _ en lugar de * y ?. En el dise√±ador de Access (propiedades de la BD), puedes verificar/cambiar este ajuste.
Nulos: Los valores Null no coinciden con Like. Si quieres excluir nulos expl√≠citamente, a√±ade AND Campo Is Not Null.
Espacios: Si el nombre del campo tiene espacios, usa corchetes [Ship Country].
Secuencias vs letras sueltas:

Like "*sw*" requiere la secuencia consecutiva sw.
Like "S*W*" asegura que exista una S y despu√©s una W, con cualquier cosa en medio.
Para ‚Äúcontiene S o W‚Äù usa Like "*S*" OR Like "*W*".


## **Criterios con campos ‚ÄúLookup‚Äù y error **Data type mismatch** en consultas de Access **

**Idea clave:** En Access, los campos **Lookup** (desplegables) muestran texto amigable (p. ej., *Speedy Express*), pero **almacenan** el **ID num√©rico** (p. ej., `1`). Si filtras con el texto en un campo que guarda n√∫meros, aparece el error **‚ÄúData type mismatch in criteria expression‚Äù**. La soluci√≥n es filtrar por el **ID** o **unir** la tabla relacionada y filtrar por el **texto** all√≠.

---

## Por qu√© ocurre el error
- El campo **ShipVia** (en `Orders`) es un **Lookup** ligado a `Shippers.ShipperID`.
- Aunque ves ‚ÄúFederal Shipping / Speedy Express / United Package‚Äù, el valor **real** guardado en `Orders.ShipVia` es **num√©rico** (1, 2, 3‚Ä¶).
- Si escribes en el criterio `‚ÄúSpeedy Express‚Äù` (con comillas), Access lo trata como **Texto** y choca con un campo **Num√©rico** ‚Üí **Data type mismatch**.

---

## Soluci√≥n 1 (r√°pida): filtrar por **ID** (sin comillas)
1. Abre la tabla **Shippers** para ver los IDs:
   - `1 = Speedy Express`
   - `2 = United Package`
   - `3 = Federal Shipping`
2. En la **Vista de dise√±o** de la consulta, en el campo **ShipVia**, escribe el ID **sin comillas**: 1

3. Ejecuta: ver√°s solo los registros de **Speedy Express**.

**SQL de ejemplo:**
```sql
SELECT OrderID, OrderDate, ShipVia
FROM Orders
WHERE ShipVia = 1;  -- 1 = Speedy Express
